```python
import yaml, os

with open("../data_sea_turtles.yaml") as f:
    data = yaml.safe_load(f)

root = os.path.abspath(data["path"])
print("Dataset root:", root)

for k in ["train", "val", "test"]:
    p = os.path.join(root, data[k])
    print(k, "exists:", os.path.exists(p), p)
```
