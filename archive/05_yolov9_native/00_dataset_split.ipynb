# Criar split train / val / test


```python
import os
import shutil
import random
from pathlib import Path

# Paths
BASE_PATH = Path("../Dataset/sea-turtles-1")
ALL_IMAGES = BASE_PATH / "all/images"
ALL_LABELS = BASE_PATH / "all/labels"


print("Images path exists:", ALL_IMAGES.exists())
print("Labels path exists:", ALL_LABELS.exists())

# Output paths
for split in ["train", "valid", "test"]:
    (BASE_PATH / split / "images").mkdir(parents=True, exist_ok=True)
    (BASE_PATH / split / "labels").mkdir(parents=True, exist_ok=True)

# Split ratios
train_ratio = 0.7
val_ratio = 0.2
test_ratio = 0.1

images = list(ALL_IMAGES.glob("*.jpg"))
random.shuffle(images)

n_total = len(images)
n_train = int(train_ratio * n_total)
n_val = int(val_ratio * n_total)

splits = {
    "train": images[:n_train],
    "valid": images[n_train:n_train+n_val],
    "test": images[n_train+n_val:]
}

for split, imgs in splits.items():
    for img_path in imgs:
        label_path = ALL_LABELS / f"{img_path.stem}.txt"

        shutil.copy(img_path, BASE_PATH / split / "images" / img_path.name)
        shutil.copy(label_path, BASE_PATH / split / "labels" / label_path.name)

print("Dataset split completed:")
print({k: len(v) for k, v in splits.items()})
```
